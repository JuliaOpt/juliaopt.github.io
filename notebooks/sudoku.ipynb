# Sudoku

![Sudoku](http://upload.wikimedia.org/wikipedia/commons/f/ff/Sudoku-by-L2G-20050714.svg)

The rules of Sudoku are

* The numbers 1 to 9 must appear in each 3x3 square
* The numbers 1 to 9 must appear in each row
* The numbers 1 to 9 must appear in each column
* And, of course, each number can only appear once in a cell.

This isn't an optimization problem, its a feasibility problem: we wish to find a feasible solution that satsifies these rules. You can think of it as an optimization problem with an objective of 0.

````julia
using JuMP
````

We will define a binary variable (a variable that is either 0 or 1) for each possible number in each cell, i.e.
```
x[i,j,k] = 1  if cell (i,j) has number k inside
```

````julia
sudoku = Model()
@defVar(sudoku, x[1:9,1:9,1:9], Bin)
````

Now we need to add constraints. Let us begin with the simplest rule: only one number per cell.

````julia
for i = 1:9, j = 1:9  # Each row and each column
    @addConstraint(sudoku, sum{x[i,j,k],k=1:9} == 1)
end
````

````julia
sudoku
````

The constraints that say each number appears once and once only in each row are very similar, so much so that we can actually share loops.

````julia
for ind = 1:9  # Each row, OR each column
    for k = 1:9  # Each digit
        # Sum across columns - row constraint
        @addConstraint(sudoku, sum{x[ind,j,k],j=1:9} == 1)
        # Sum across rows - column constraint
        @addConstraint(sudoku, sum{x[i,ind,k],i=1:9} == 1)
    end
end
````

````julia
sudoku
````

A trickier constraint to index is the 3x3 square constraint. What we will do is to index over the top-left corners of each 3x3 square with `for` loops, then sum over the square.

````julia
for i = 1:3:7, j = 1:3:7, k = 1:9
    @addConstraint(sudoku, sum{x[r,c,k], r=i:i+2, c=j:j+2} == 1)
end
````

The final step is to add the initial solution as a set of constraints. We'll solve the problem at the start of this notebook.

````julia
init_sol = [ 5 3 0 0 7 0 0 0 0;
             6 0 0 1 9 5 0 0 0;
             0 9 8 0 0 0 0 6 0;
             8 0 0 0 6 0 0 0 3;
             4 0 0 8 0 3 0 0 1;
             7 0 0 0 2 0 0 0 6;
             0 6 0 0 0 0 2 8 0;
             0 0 0 4 1 9 0 0 5;
             0 0 0 0 8 0 0 7 9]
for i = 1:9, j = 1:9
    if init_sol[i,j] != 0
        @addConstraint(sudoku, x[i,j,init_sol[i,j]] == 1)
    end
end
````

````julia
solve(sudoku)
````

To display the solution, we need to look for the values of x[i,j,k]
that are 1.

````julia
x_val = getValue(x)
sol = zeros(Int,9,9)
for i in 1:9, j in 1:9, k in 1:9
    # LP/IP solvers are not precise so even binary variables
    #  may be slightly less than 1.
    if x_val[i, j, k] >= 0.9  
        sol[i, j] = k
    end
end
sol
````
