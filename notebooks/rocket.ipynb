# Rocketry control problem

* Maximize: final altitude
* Control: thrust
* States: velocity, altitude, mass of fuel and rocket
* Dynamics: gravity, thrust, drag forces
* Use dimensionless data

http://www.mcs.anl.gov/~more/cops/cops3.pdf

```julia
using JuMP, Ipopt
```

```julia
# Constants
h_0 = 1    # Initial height
v_0 = 0    # Initial velocity
m_0 = 1    # Initial mass
g_0 = 1    # Gravity at the surface

# Parameters
T_c = 3.5
h_c = 500  # Used for drag
v_c = 620  # Makes things dimensionless
m_c = 0.6  # Fraction of initial mass

# Derived parameters
c     = 0.5*sqrt(g_0*h_0)  # Thrust-to-fuel mass
m_f   = m_c*m_0            # Final mass
D_c   = 0.5*v_c*m_0/g_0    # Drag scaling
T_max = 3.5*g_0*m_0        # Maximum thrust

n   = 1000    # Time steps
Δt  = 0.2/n     # Time step
t_f = Δt*n    # Time of flight

mod = Model(solver=IpoptSolver())
@defVar(mod, v[0:n] ≥ 0)  # Velocity
@defVar(mod, h[0:n] ≥ h_0)  # Height
@defVar(mod, m_f ≤ m[0:n] ≤ m_0)  # Mass

# Objective: maximize altitude at end of time of flight
@setNLObjective(mod, Max, h[n])

# Control: thrust
@defVar(mod, 0 ≤ T[0:n] ≤ T_max)

# Initial conditions
@addConstraint(mod, v[0] == v_0)
@addConstraint(mod, h[0] == h_0)
@addConstraint(mod, m[0] == m_0)
@addConstraint(mod, m[n] == m_f)

# Dynamics
for j in 1:n
    # h' = v
    @addNLConstraint(mod, h[j] == h[j-1] + Δt*v[j-1])

    # v' = (T-D(h,v))/m - g(h)
    # D(h,v) = D_c*v²*exp(-h_c*(h-h(0))/h(0))
    # g(h) = g_0*(h(0)/h)^2
    @addNLConstraint(mod, v[j] == v[j-1] + 
                    Δt*(T[j-1]
                        - D_c*(v[j-1]^2)*exp(-h_c*(h[j-1]-h_0)/h_0) 
                        - m[j-1]*g_0*(h_0/h[j-1])) / m[j-1])

    # m' = -T/c
    @addNLConstraint(mod, m[j] == m[j-1] - Δt*T[j-1]/c)
end

# Provide starting solution
for k in 0:n
    setValue(h[k], 1)
    setValue(v[k], (k/n)*(1 - (k/n)))
    setValue(m[k], (m_f - m_0)*(k/n) + m_0)
    setValue(T[k], T_max/2)
end
println("Solving...")
status = solve(mod)
@show status
```

```julia
np = n #div(n,4)
h_plot = plot(x=(0:np)*Δt,y=getValue(h)[1:np+1], Geom.line,
                Guide.xlabel("Time (s)"), Guide.ylabel("Altitude"))
m_plot = plot(x=(0:np)*Δt,y=getValue(m)[1:np+1], Geom.line,
                Guide.xlabel("Time (s)"), Guide.ylabel("Mass"))
v_plot = plot(x=(0:np)*Δt,y=getValue(v)[1:np+1], Geom.line,
                Guide.xlabel("Time (s)"), Guide.ylabel("Velocity"))
T_plot = plot(x=(0:np)*Δt,y=getValue(T)[1:np+1], Geom.line,
                Guide.xlabel("Time (s)"), Guide.ylabel("Thrust"))
draw(SVG(6inch, 6inch), vstack( hstack(h_plot,m_plot),
                                hstack(v_plot,T_plot)))
```
